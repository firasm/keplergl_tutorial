```python
import pandas as pd
import geopandas as gpd

import keplergl
```

## Canada Census Map

```python
canada = gpd.read_file('canada_cd_sim.geojson')
canada['CDUID'] = canada['CDUID'].astype(int)
census = pd.read_csv('wrangled_census.csv')
census = census[census['Year']==2016]
```

```python
canada
```

```python
full = canada.merge(census, right_on = 'geographic_code', left_on='CDUID')

full
```

```python
full['geometry'][5]
```

```python
print(full['geometry'][5])
```

```python
full["coord-x"] = full['geometry'].centroid.map(lambda p: p.x)
full["coord-y"] = full['geometry'].centroid.map(lambda p: p.y)
```

```python
w2 = keplergl.KeplerGl(height=800, width=600)
w2.add_data(full, name = "Census")
w2
```

```python
w3 = keplergl.KeplerGl(height=1000)
w3.add_data(canada, name = "Regions")
w3.add_data(census, name = 'census')
w3
```
